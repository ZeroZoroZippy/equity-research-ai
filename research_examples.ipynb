# üöÄ Enhanced Equity Research System

## Two Powerful Modes:
1. **Single Company Research** - Deep dive into one stock
2. **Sector Research** - Find and compare top companies in a sector

````python
# Import the enhanced system
from research_system import EquityResearchSystem
from IPython.display import display, Markdown
from dotenv import load_dotenv

load_dotenv()
````

````python
# Create the research system
system = EquityResearchSystem()
print("‚úÖ Research system initialized!")
````

---

## üîç Mode 1: Single Company Research

**Use this when:** You know which stock you want to research

````python
# Example 1: Research Apple (US market)
report = await system.research_stock("AAPL", "US")
display(Markdown(report))
````

````python
# Example 2: Research an Indian company (NSE)
report = await system.research_stock("RELIANCE", "NSE")
display(Markdown(report))
````

---

## üåê Mode 2: Sector Research

**Use this when:** You want to invest in a sector but don't know which company

**This mode will:**
1. Identify top 5 companies in the sector
2. Research EACH company in detail (using Mode 1)
3. Compare all companies
4. Recommend the best investment with portfolio allocation

````python
# Example 3: Research Technology Sector (top 3 companies)
# ‚ö†Ô∏è This will take 15-20 minutes (researching 3 companies + comparison)
sector_report = await system.research_sector(
    sector="Technology",
    exchange="US",
    num_companies=3  # Start with 3 companies for faster results
)

display(Markdown(sector_report))
````

````python
# Example 4: Research Indian Banking Sector
banking_report = await system.research_sector(
    sector="Banking",
    exchange="NSE",
    num_companies=4
)

display(Markdown(banking_report))
````

---

## üíæ Save Reports to Files

````python
# Save single company report
report = await system.research_stock("TSLA", "US")
with open("TSLA_research.md", "w") as f:
    f.write(report)
print("‚úÖ TSLA report saved to TSLA_research.md")
````

````python
# Save sector report
sector_report = await system.research_sector("Healthcare", "US", 3)
with open("healthcare_sector.md", "w") as f:
    f.write(sector_report)
print("‚úÖ Healthcare sector report saved to healthcare_sector.md")
````

---

## üéØ Quick Comparison: Different Sectors

````python
# Compare different sectors to see which looks more promising

print("üîç Researching Technology sector...")
tech = await system.research_sector("Technology", "US", 2)

print("üîç Researching Healthcare sector...")
healthcare = await system.research_sector("Healthcare", "US", 2)

print("\n" + "="*60)
print("TECHNOLOGY SECTOR")
print("="*60)
display(Markdown(tech))

print("\n" + "="*60)
print("HEALTHCARE SECTOR")
print("="*60)
display(Markdown(healthcare))

print("\n‚úÖ Now compare both sectors and decide which looks better!")
````
